### Import Library

```python
#import seluruh library yang akan digunakan
import pandas as pd
import requests
from sqlalchemy import create_engine
```

### Koneksi ke Detail of SIMDASI Master Table

```python
# URL untuk mendapatkan tabel referensi tabel dari SIMDASI
detailtabel_url = "https://webapi.bps.go.id/v1/api/interoperabilitas/datasource/simdasi/id/34/key/7342305c0d7d5f9fe5a2f0e4c7d08e57"
req_detailtabel = requests.get(detailtabel_url).json()

# Jika API memerlukan parameter "key", tambahkan ke URL sesuai dengan kebutuhan
# detailtabel_url = "https://webapi.bps.go.id/v1/api/interoperabilitas/datasource/simdasi/id/34/key/YOUR_KEY/token"
# req_detailtabel = requests.get(detailtabel_url).json()

# Normalisasi data JSON menggunakan pandas.json_normalize
detailtabel = pd.json_normalize(req_detailtabel['data'][1]['data'])

# Mendapatkan kolom 'kode' sebagai list
# id_tabel = list(detailtabel['id_tabel'])
# list_detailtabel = detailtabel.values.tolist()
df_detailtabel = pd.DataFrame(detailtabel)
df_detailtabel
```

### Import ke dalam database

```python
# Ganti dengan informasi koneksi database Anda
database_url = 'postgresql://postgres:password@localhost:5432/produksi'
engine = create_engine(database_url)
nama_tabel =  'master_table_SIMDASI'
nama_skema = 'statistik_bps_simdasi'

# Simpan DataFrame ke dalam tabel database dengan skema
df_detailtabel.to_sql(nama_tabel, engine, schema=nama_skema, if_exists='replace', index=False)

print(f'Tabel {nama_tabel} berhasil dimuat ke dalam database dengan skema {nama_skema}.')
```
